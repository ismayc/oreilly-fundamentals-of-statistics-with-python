## Walkthrough 3.2: t-tests

> **Common Pitfall:** A p-value is not the probability that the null hypothesis is true. It is the probability of seeing data at least as extreme as yours if the null hypothesis were true. A small p-value is evidence against the null, not a statement about how likely the null is.

> **Common Pitfall:** The `alternative` argument controls whether the test is one-tailed (`'greater'` or `'less'`) or two-tailed (`'two-sided'`, the default). Pick the direction before you see the data, based on the question. A one-tailed test only detects an effect in the specified direction, so using the wrong tail can hide a real effect.

## Walkthrough 3.4: Non-Parametric Tests

> **Common Pitfall:** Non-parametric tests like Mann-Whitney U and Kruskal-Wallis do not assume normally distributed data, which makes them useful for skewed data such as salaries. But they compare distributions (ranks/medians) rather than means, so do not report them as tests of the difference in means. Reach for them when the normality assumption of a t-test or ANOVA is clearly violated.
